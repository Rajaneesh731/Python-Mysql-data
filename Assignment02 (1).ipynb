```python
#Exporting the csv into elasticsearch database
from elasticsearch import helpers, Elasticsearch
import csv

es = Elasticsearch(['101.53.136.181'],port=9200)     #es cluster of given host

with open("/home/rajaneesh/Downloads/beginner_assignment01 - product_listing.csv") as f:    #getting excel data
    reader = csv.DictReader(f)
    helpers.bulk(es, reader, index='rajaneesh', doc_type='my-type')      
```
